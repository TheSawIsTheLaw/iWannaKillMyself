```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("csv.csv")
month = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]
```

```python
from cycler import cycler

monochrome = (cycler('color', ['k']) * cycler('linestyle', ['--', ':']) * cycler('marker', ['^',',', '.']))

fig, ax = plt.subplots(1,1)
ax.set_prop_cycle(monochrome)

ax.plot(month, df["2018"], label="2018")
ax.plot(month, df["2019"], label="2019")
ax.plot(month, df["2020"], label="2020")
ax.plot(month, df["2021"], label="2021")
ax.plot(month, df["2022 (Provisional)"], label="2022")

plt.xticks(rotation = -45, ha="left", rotation_mode="anchor")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), shadow=True)
plt.ylabel("Количество смертей")
plt.xlabel("Месяц")
ax.grid(axis="y", linestyle= "--")
fig.savefig("deathToMonth.pdf", format="pdf",bbox_inches="tight")
fig.show()
```
